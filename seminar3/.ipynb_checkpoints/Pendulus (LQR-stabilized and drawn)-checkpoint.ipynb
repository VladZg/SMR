```python
import numpy as np
from math import sqrt, sin, cos
import cv2
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
from scipy import linalg

def plot_1d(data):
    plt.figure(figsize=(15, 4))
    plt.plot(data)
    plt.show()
```

```python
class Pendulus:
    def __init__(self, m = 5.0, l = 1.0, k = 15.0, g = 10.0,
                 phi1_0 = 0.0, phi2_0 = 0.0, w1_0 = 0.0, w2_0 = 0.0,
                 dt = 0.001, Q = np.eye(4), R = np.eye(1)):
        
        self.x  = np.array([[phi1_0], [w1_0], [phi2_0], [w2_0]])
        # print(self.x, self.x.T)
        
        self.m  = m
        self.l  = l
        self.g  = g
        self.k  = k
        self.dt = dt
        
        self.A = np.array\
        ([[0,                          1,     0,                          0],
          [-(m * g + k * l) / (m * l), -0.02, k / m,                      0],
          [0,                          0,     0,                          1],
          [k / m,                      0,     -(m * g + k * l) / (m * l), -0.02]])
        
        self.B = np.array([[0], [0], [0], [1.0 / (m * l)]])
        
        self.Q = Q
        self.R = R

        # solving ARE
        self.P = linalg.solve_continuous_are(self.A, self.B, self.Q, self.R)
        # LQR feedback
        self.K = np.linalg.inv(R) @ self.B.T @ self.P
        
    def get_state(self):
        #print("state shape", self.x.shape)
        
        return self.x

    # вычисление динамики системы по заданному управлению
    # f(x,u) = A*x + B*u
    def sys_dyn(self, u):
        x_dot = np.zeros_like(self.x)
        
        x_dot = self.A @ self.x + self.B * u
        
        return x_dot

    # x_(k+1) = x_k + x_dot_k * dt
    def integrate_sys_dyn(self, sys_dyn):
        self.x = self.x + sys_dyn * self.dt
    
    def lqr_feedback(self, x):

        # K = np.array([[-1.0],[-sqrt(3.0)]])
        u = - self.K @ x
        
        return u
```

```python
def run_episode(system, iter_num = 75000):
    i = 0

    x_traj = []
    u_traj = []
    
    while (i < iter_num):
        state = system.get_state()

        # control = np.array([[0]])
        control = system.lqr_feedback(state)

        x_dot = system.sys_dyn(control)
        #print("xdot: ", x_dot)
        
        system.integrate_sys_dyn(x_dot)
        
        x_traj.append(state)
        u_traj.append(control)

        i += 1
    
    return x_traj, u_traj
```

```python
def state_action_cost(x, u, Q, R):
    # print(x.shape, u.shape)
    # print(x, x.T)
    # print(u)
    # print(Q)
    # print(R)
    
    cost = x.T @ Q @ x + u.T @ R @ u
    
    return cost

def episode_cost(x_hist, u_hist, Q, R):
    total_cost = 0
    cost_hist = []
    
    for x, u in zip(x_hist, u_hist):
        #print(x)
        #print(v)
                
        cost = state_action_cost(x, u, Q, R)
        
        total_cost += cost
        cost_hist.append(cost)
    
    return total_cost, cost_hist
```

```python
def run_and_plot_episode(system, iter_num = 75000):    
    x_hist, u_hist = run_episode(system, iter_num)
    total_cost, cost_hist = episode_cost(x_hist, u_hist, system.Q, system.R)

    m = system.m
    l = system.l
    g = system.g
    k = system.k
    
    phi1_list = [x[0, 0] for x in x_hist]
    w1_list = [x[1, 0] for x in x_hist]
    phi2_list = [x[2, 0] for x in x_hist]
    w2_list = [x[3, 0] for x in x_hist]
    u_hist_list = [u[0, 0] for u in u_hist]
    cost_hist_list = [cost[0, 0] for cost in cost_hist]
    energy_hist_list = [(m*g*l*(2.0-cos(x[0,0])-cos(x[2,0]))+0.5*k*(l**2)*(sin(x[0,0])-sin(x[2,0]))**2+0.5*m*(l**2)*(x[1,0]**2+x[3,0]**2)) for x in x_hist]
    
    plot_1d(phi1_list)
    plot_1d(w1_list)
    plot_1d(phi2_list)    
    plot_1d(w2_list)
    plot_1d(u_hist_list)
    plot_1d(cost_hist_list)
    plot_1d(energy_hist_list)

    plt.figure(figsize=(15, 15))
    #plt.plot(x_hist, v_hist)
    plt.show()
    # clear_output(wait = True)
    
    return x_hist, u_hist
```

```python
system = Pendulus(phi1_0 = 0.1, w1_0 = 1.0, phi2_0 = -0.1, w2_0 = 0.3, Q = np.eye(4), R = np.eye(1) * 0.1)

run_and_plot_episode(system, 75000)
```

```python
def run_and_plot_episode_wrapper(k = 15.0, r = 0.1):
    system = Pendulus(k = k, phi1_0 = 0.1, w1_0 = 1.0, phi2_0 = -0.1, w2_0 = 0.3, Q = np.eye(4), R = np.eye(1) * r)
    run_and_plot_episode(system, 75000)

# how graphics depend on parameters
widgets.interact(run_and_plot_episode_wrapper, k=widgets.IntSlider(min=0.0, max=50.0, step=0.5, value=15.0), r=widgets.IntSlider(min=0.1, max=10.0, step=0.1, value=1))
```

```python
def draw_system(canvas, cr, x, l, t, color = (234, 23, 123)):
    if (canvas is None):
        canvas = np.ones((700, 700, 3)) * 0
    
    h, w, _ = canvas.shape

    phi1 = x[0,0]
    phi2 = x[2,0]

    x0, y0 = w//2, h//2
    x1, y1 = x0 + l * sin(phi1), y0 + l * cos(phi1)
    x2, y2 = x0 + l * sin(phi2), y0 + l * cos(phi2)

    spring_len = sqrt((x1-x2)**2+(y1-y2)**2)
    spring_len_norm = int(1 + 10 - 10 * (spring_len / (2.0*l)))
    
    x0, y0 = int(x0), int(y0)
    x1, y1 = int(x1), int(y1)
    x2, y2 = int(x2), int(y2)

    clr1 = (144, 20, 222)
    clr2 = (255, 153, 0)
    white_clr = (255, 255, 255)
    
    cv2.circle(canvas, (x0,y0), 0, white_clr, 2)
    cv2.circle(canvas, (x1, y1), cr, clr1, 2)
    cv2.circle(canvas, (x2, y2), cr, clr2, 2)
    cv2.line(canvas, (x0,y0), (x1,y1), white_clr, 4) 
    cv2.line(canvas, (x0,y0), (x2,y2), white_clr, 4) 
    cv2.line(canvas, (x1,y1), (x2,y2), white_clr, spring_len_norm, lineType=4) 

    font = cv2.FONT_HERSHEY_SIMPLEX 
    cv2.putText(canvas,'phi1 = {} grad'.format(round(x[0,0], 2)),(20,30), font, 1, clr1, 1)
    cv2.putText(canvas,'w1  = {} grad/sec'.format(round(x[1,0], 2)),(20,60), font, 1, clr1, 1)
    cv2.putText(canvas,'phi2 = {} grad'.format(round(x[2,0], 2)),(20,90), font, 1, clr2, 1)
    cv2.putText(canvas,'w2  = {} grad/sec'.format(round(x[3,0], 2)),(20,120), font, 1, clr2, 1)
    cv2.putText(canvas,'t    = {} sec'.format(round(t, 1)),(20,150), font, 1, white_clr, 1)

    
    return canvas
```

```python
def run_and_draw_episode(system, iter_num = 75000, step = 10):
    x_hist, u_hist = run_episode(system, iter_num)

    canvas = np.ones((700, 700, 3), np.uint8) * 70
    
    i = 0
    t = 0.0
    i_max = len(x_hist)
    
    while (True):
        x = x_hist[i]
        # print(x, i, t)        
        
        canvas[:, :, :] = 70
        canvas = draw_system(canvas, 30, x, system.l, t)
        cv2.imshow("Pendulus", canvas)

        i += step
        t += step*system.dt

        if (i > i_max):
            break

        key = cv2.waitKey(10) & 0xFF
        if (key == ord('q')):
            break

    cv2.waitKey(0)
    cv2.destroyAllWindows()
    cv2.waitKey(10)
```

```python
system = Pendulus(k = 0.5, l = 100.0, phi1_0 = -0.3, w1_0 = -1.0, phi2_0 = 0.2, w2_0 = 0.7, Q = np.eye(4), R = np.eye(1) * 0.1)

# run_and_plot_episode(system, iter_num = 175000)
run_and_draw_episode(system, iter_num = 175000, step = 20)
```
